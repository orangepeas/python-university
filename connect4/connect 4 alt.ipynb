```python
###### plot gameboard
#loop over even and odd numbers to check player turns
#ask for an input for where you want the token
#check if the input is a legal move
#plot the arrayboard on the gameboard
#check if there is 4 in a row after the input, and then if the board is full, if not next player turn

#TO MAYBE ADD: make it so that move inputted is just a column, since row doesnt matter to player, add docstrings to 
#all the functions
import numpy as np
from matplotlib import pyplot as plt
import sys

def playerturn(i,graphicboard):
    """
    main loop for playervsplayer, plots the arrayboard, asks player to input move, plots the move on the arrayboard
    then plots the move on the graphics board, checks if the game has been won and if it has then asks if they want
    to play again
    """
    arrayboard=np.zeros([6,7])
    gamewon=False                         
    while gamewon==False:
        if i%2==0:            #if the number is even, it is player 1's turn
            playerone=True
        else:
            playerone=False
        playermove=move(playerone,arrayboard)   #asks the player to input move
        arrayboard=plotmove(playermove,playerone,arrayboard)
        graphicalboard(playermove,arrayboard,playerone,graphicboard)
        (gamewon,winner)=gamewin(arrayboard)
        i=i+1
    if gamewon==True:
        if winner==True:
            p1="Player One"
        else:
            p1="Player Two"
        print("Congratulations ",p1,", you have won Connect 4!")
        print()
        playagain()
            
            
def playerturn2(i,graphicboard):
    """
    same as playerturn except for playervscomputer,
    player two's move (the computer's move), is executed by the function robomove()
    """
    arrayboard=np.zeros([6,7])
    gamewon=False                         
    while gamewon==False:
        if i%2==0:            #if the number is even, it is the human's turn
            playerone=True
            playermove=move(playerone,arrayboard)   #asks the player to input move
            arrayboard=plotmove(playermove,playerone,arrayboard)
            graphicalboard(playermove,arrayboard,playerone,graphicboard) 
            (gamewon,winner)=gamewin(arrayboard)                          
        else:
            robomove(arrayboard,graphicboard)
        i=i+1
    if gamewon==True:
        p1="Player One"
        print("Congratulations ",p1,",","you have won Connect 4!")
        print()
        playagain()
            
            

def plotmove(move,playerone,arrayboard):
    """
    plots the inputted move on the arrayboard
    """
    if playerone==True:
        arrayboard[move[0],move[1]]=1
        #print(arrayboard)
        return(arrayboard)
    else:
        arrayboard[move[0],move[1]]=2
        #print(arrayboard)
        return(arrayboard)

        

def graphicalboard(move,arrayboard,playerone,board):
    """
    plots the inputted move on the graphics board, first by converting the coordinates into graph coordinates,
    then plotting the player's moves with their respective colours depending on which player inputted
    """
    for i in range(arrayboard.shape[0]):           #switches to a graph coordinate system
        if move[0]==i:                             #loops to see what the move is and mirrors in x axis 
            move[0]=(arrayboard.shape[0]-i)
            break
        else:
            continue
    xmove=(move[1]+1)                            
    ymove=(move[0])
    if playerone==True:   
        plt.plot(xmove,ymove,'ro',markersize=11)
    else:
        plt.plot(xmove,ymove,'o',c='yellow',markersize=11)
    plt.show(False) #IT WORKS 

        
        
def move(playeroneturn,arrayboard):
    """
    asks the player to input a move, and then passes it the the check() function to see if it's legal, and if it's
    legal it returns the move into the main playerturn loop, which then passes the move into the plot function for
    both the arrayboard and the graphics board
    """
    failstate=False
    if playeroneturn == True: #checks to see which players turn it is
        p=("Player One, ")
    else:
        p=("Player Two, ")
    while True: #takes the input and checks if its legal, and then passes to the plot function

        try:
            print()        
            print(p)
            print()
            move=eval(input("Please enter the co-ordinates of where you would like your token in an x-y format: "))
        except:
            print("Please enter a legal move.")
            continue
        (correctmove,failstate)=check(move,arrayboard) 
        if failstate==True:
            continue
        else:
            return(correctmove)
        

            
def check(move,arrayboard): 
    """
    first it converts the move into an array, with correct matrix denotation instead of x-y coordinate denotation 
    checks if the inputted move is within the bounds of the board, if its a tuple, if its a legal connect 4 move
    """
    failstate=False
    if type(move)==tuple and len(move)==2 and move[0]<=(arrayboard.shape[1]) and move[1]<=(arrayboard.shape[0]):
        move=(np.array(move)-1)                         #minuses 1 because computer starts at 0
        move[[0,1]]=move[[1,0]]            #swaps variables so that coordinates are x,y instead of y,x
        for i in range(arrayboard.shape[0]):           #switches to the coordinate system that i want
            if move[0]==i:                             #loops to see what the move is and mirrors in x axis 
                move[0]=(arrayboard.shape[0]-i)-1
                break
            else:
                continue
        a=arrayboard[(move[0]),move[1]]           #it doesnt like if this is the if comparison without the variable
        if a!=0:                                #checks if there is a token where the move is
            print("Please move where there is no token.")
            failstate=True                
            return(move,failstate)

        if move[0]<0 or move[1]<0:                                #checks if there is a token where the move is
            print("Please enter a legal move.")
            failstate=True                
            return(move,failstate)
        (move,failstate)=legalmove(move,arrayboard,failstate)              
        if failstate == True:
            return(move,failstate)
        
        failstate=False
        return(move,failstate)
        
    else:
        print("Please enter a legal move.")
        failstate=True              
        return(move,failstate)

def gamewin(arrayboard):  
    """
    checks if the game has been won by looping over columns, rows, left ascending diagonals and right ascending 
    diagonals, to see if 4 in a row has been won by either player
    """
    p1=83        #initialising it to something that doesnt matter so python doesnt cry when it is returned
    ysize=(arrayboard.shape[0]-1)       #5
    xsize=(arrayboard.shape[1]-1)       #6
    for i in range(xsize):
        for j in range(ysize-1):       #checks rows, -1 bcs it checks 3 in front, only needs to check 4 times
            if arrayboard[i][j]==arrayboard[i][j+1]==arrayboard[i][j+2]==arrayboard[i][j+3]==1: 
                p1=True
                gamewon=True
                return(gamewon,p1)     

        for j in range(ysize-1):                             #checks rows for p2
            if arrayboard[i][j]==arrayboard[i][j+1]==arrayboard[i][j+2]==arrayboard[i][j+3]==2: 
                p1=False
                gamewon=True
                return(gamewon,p1) 
                
    for j in range(xsize+1):
        for i in range(xsize-3):                             #checks columns
            if arrayboard[i][j]==arrayboard[i+1][j]==arrayboard[i+2][j]==arrayboard[i+3][j]==1: 
                p1=True
                gamewon=True
                return(gamewon,p1) 

        for i in range(xsize-3):                             #checks columns for p2
            if arrayboard[i][j]==arrayboard[i+1][j]==arrayboard[i+2][j]==arrayboard[i+3][j]==2: 
                p1=False
                gamewon=True
                return(gamewon,p1) 
    
    for j in range(xsize-2):

        for i in range(ysize-2):                             #checks rightward ascending diagonals

            if arrayboard[i][j]==arrayboard[i+1][j+1]==arrayboard[i+2][j+2]==arrayboard[i+3][j+3]==1: 
                p1=True
                gamewon=True
                return(gamewon,p1)     

        for i in range(ysize-2):                             #checks rght asc diags for p2
            if arrayboard[i][j]==arrayboard[i+1][j+1]==arrayboard[i+2][j+2]==arrayboard[i+3][j+3]==2: 
                p1=False
                gamewon=True
                return(gamewon,p1)   
            
    for j in range(xsize-2):
        #print("j is",j)
        #print("j+3 is",j+3)
        for i in range(ysize,2,-1):                             #checks leftward ascending diagonals
            #print("i is",i)
            #print("i-3 is",i-3)   
            if arrayboard[i][j]==arrayboard[i-1][j+1]==arrayboard[i-2][j+2]==arrayboard[i-3][j+3]==1: 
                p1=True
                gamewon=True
                return(gamewon,p1)     

        for i in range(ysize,2,-1):                             #checks lft asc diags for p2
            if arrayboard[i][j]==arrayboard[i-1][j+1]==arrayboard[i-2][j+2]==arrayboard[i-3][j+3]==2: 
                p1=False
                gamewon=True
                return(gamewon,p1) 
        if np.all(arrayboard)!=0:
            gamedraw()
            
    gamewon=False
    return(gamewon,p1)
    
    
def plottedmove(move,arrayboard):
    """
    this is a copy paste of the x,y coordinates into matrix coordinates section of the check code,
    wasn't used for a reason i cant remember, i think i wanted to put this section into a function but it didnt work
    for some reason
    """
    move=(np.array(move)-1)              #minuses 1 because computer starts at 0
    for i in range(arrayboard.shape[0]):           #switches to the coordinate system that i want
        if move[0]==i:                             #loops to see what the move is and mirrors in x axis 
            move[0]=(arrayboard.shape[0]-i)-1
            break
        else:
            continue
    return(move)

def legalmove(move,arrayboard,failstate):
    """
    checks if the move is on the bottom row or there's one below it, if there isn't, failstate is true and illegal move
    """
    failstate==False
    x=move[0]
    y=move[1]                        #again it doesnt like comparisons without a neat variable so  here i am
    if x==5:                         #1st row, coordinates are flipped and start at 0 
        failstate==False
        return(move,failstate)
    elif arrayboard[x+1,y]==0:      #plus instead of minus since coordinates are flipped such that 6 is below 5
        print("Please enter a legal move.")
        failstate=True                
        return(move,failstate)
    else:
        failstate==False
        return(move,failstate)

    
    
def robomove(arrayboard,graphicboard):
    """
    the main loop function for the computer's move, first the computer randomly generates moves and checks them until
    it finds a legal move, then it plots it on the arrayboard, the graphics board, and then checks if the game has been
    won by the robot, and if it has it prints some insults
    """
    playerone=False
    clegalmove=True
    while clegalmove==True:    #if its true it means there has been a failstate
        import random as rand
        x=rand.randrange(6)
        y=rand.randrange(7)
        cmove=[x,y]                                      #I didnt want to rewrite
        (null,clegalmove)=robocheck(cmove,arrayboard)    #the check function, i just copy pasted it. It returns
    arrayboard=plotrobomove(cmove,arrayboard)            #a move because usually it corrects co-ordinates
    graphicalboard(cmove,arrayboard,playerone,graphicboard) #but that isnt necessary here so it is stored in null
    (gamewon,winner)=gamewin(arrayboard)               
    if gamewon==True:
        print("you lost against the robot placing tokens randomly")
        print("you bring shame to your family")
        print("how embarassing...")
        playagain()
    

    
def plotrobomove(move,arrayboard):
        """
        same as plotmove it just doesn't check what player it is since it knows its playertwo, could've just told the 
        plotmove function that the move was made by player twp instead of making an entirely new function but
        oh well
        """
        arrayboard[move[0],move[1]]=2
        #print(arrayboard)
        return(arrayboard)

    

def robocheck(move,arrayboard): 
    """
    same as check except it doesn't need to convert co-ordinates from x-y since they were chosen randomly and not
    inputted as x-y
    """
    if type(move)==list and len(move)==2 and move[0]<=(arrayboard.shape[1]) and move[1]<=(arrayboard.shape[0]):
        failstate=False
        a=arrayboard[(move[0]),move[1]]           #it doesnt like if this is the if comparison without the variable
        if a!=0:                                #checks if there is a token where the move is
            failstate=True
            return(move,failstate)

        if move[0]<0 or move[1]<0:                                #checks if there is a token where the move is
            failstate=True    
            return(move,failstate)
        (move,failstate)=robolegalmove(move,arrayboard,failstate)              
        if failstate == True:
            return(move,failstate)
        failstate=False
        return(move,failstate)
    else:
        failstate=True
        return(move,failstate)
    

def robolegalmove(move,arrayboard,failstate):
    """
    checks if its on the bottom row or there's one below it, if there is failstate is true and illegal move,
    only difference between this and normal legalmove is that it doesn't print please enter a legal move
    every time there isnt one below it
    """
    failstate==False
    x=move[0]
    y=move[1]                        #again it doesnt like comparisons without a neat variable so  here i am
    if x==5:                         #1st row, coordinates are flipped and start at 0 
        failstate==False
        return(move,failstate)
    elif arrayboard[x+1,y]==0:      #plus instead of minus since coordinates are flipped such that 6 is below 5
        failstate=True                
        return(move,failstate)
    else:
        failstate==False
        return(move,failstate)

    
    
def playagainsthuman():
    """
    initialises the graphical board and the turn counter, then runs the main loop for playervsplayer, playerturn
    """
    for i in range(6):
        x=np.arange(1,8)
        y=np.full(7,[i+1])
        board=plt.scatter(x,y,c='white',s=200,edgecolors='black')
    plt.show(False)    
    i=0
    playerturn(i,board)


    
def humanorcomputer():
    """
    the first function that gets run, asks the player to input 1 or 2 for playervsplayer or playervscomputer,
    and then runs playagainsthuman or playagainstcomputer depending on what they chose
    """
    while True:        #asks the player whether they want to play against a human or a computer
        try:
            HorC=eval(input("Would you like to play against another human or a computer? 1 for human 2 for computer: "))
        except:
            print("Please enter either 1 or 2")
            continue
        if HorC==1:
            playagainsthuman()
        elif HorC==2:
            playagainstcomputer()
        else:
            print("Please enter either 1 or 2")
            continue
            
def playagain():
    """
    runs after the game has been won and asks the user if they would like to play again, and if they say no the system
    exits, if they say yes then it runs humanorcomputer() again
    """
    while True:        
        try:
            again=eval(input("Would you like to play again? Please input 1 for Yes or 2 for No: "))
        except:
            print()
            print("Please enter either 1 for Yes or 2 for No")
            continue
        if again==1:
            print()
            plt.clf()
            plt.close()
            humanorcomputer()
        if again==2:
            sys.exit(0)  #i am using sys.exit since quit() and exit() didnt do anything
            
            
def playagainstcomputer():
    """
    initialises the graphical board and the turn counter, then runs the main loop for playervscomputer, playerturn2
    """
    for i in range(6):
        x=np.arange(1,8)
        y=np.full(7,[i+1])
        board=plt.scatter(x,y,c='white',s=200,edgecolors='black',facecolor='blue')
    plt.show(False)
    i=0
    playerturn2(i,board)    

def gamedraw():
    print()
    print("The game has been drawn")
    print()
    playagain()

humanorcomputer()
```
